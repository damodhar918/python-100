```python
from urllib import response
import requests
from bs4 import BeautifulSoup


response = requests.get("https://news.ycombinator.com/news")
web_page = response.text
```

```python
response.text
```

```python
soup = BeautifulSoup(web_page, "html.parser")
article_tag = soup.find(name="a", class_="titlelink")
article_text = article_tag.getText()
article_link = article_tag.get("href")
print('article_tag: ', article_tag.getText(), article_tag.get("href"))
```

```python
soup.find(name="span", class_="score").getText()
```

```python
article_tag = soup.find_all(name="a", class_="titlelink")
article_texts =[]
article_links =[]
for art in article_tag:
    article_texts.append(art.getText())
    article_links.append(art.get("href"))    
    
article_upvotes = [int(x.text.split()[0]) for x in soup.find_all(name="span", class_="score")]
```

```python
max_ind = article_upvotes.index(max(article_upvotes))
print(article_texts[max_ind], article_links[max_ind], article_upvotes[max_ind])
```
